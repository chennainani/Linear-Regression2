```python
House_rental=pd.read_csv("D:\linear regression./attachment_house_rental_data_lyst1654.csv")
```

```python
House_rental.head()
```

```python
x=House_rental['Bedroom'].values
y=House_rental['Price'].values
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)

n=len(x)
numer=0
denom=0

for i in range (n):
    numer  += (x[i]-mean_x) * (y[i]-mean_y)  #(x-xbar)*(y-ybar)
    denom  += (x[i]-mean_x)**2   #(x-xbar)**2

    
m=numer/denom  #slope m

c=mean_y-(m*mean_x)  #intercept c
 
print('m= ',m, 'c= ',c)
  
```

```python
y_pred =[]
 
def predict(x,n):
        for i in range(n): 
            y = m*x[i] + c
            y_pred.append(y)
            
        return y_pred 
predict(x,n)
```

```python
plt.scatter(x,y, color='g')
plt.scatter(x,y_pred, color='r')
plt.plot(x,y_pred, color='m')
plt.show()
```

```python
ss_t=0 # sum of square
ss_r=0 # sum of residuals

for i in range(n):
    ss_t += (y[i]- mean_y)**2    #(y-mean_y)**2
    ss_r +=(y[i]-y_pred[i])**2  #(y-y_pred_mean)**2
r2= 1-(ss_r/ss_t)
print('r2 -score:', r2)


```
